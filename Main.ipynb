```python
from google.colab import drive
import os
from datetime import datetime

# Mount Google Drive (safe to run every time)
if not os.path.exists('/content/drive/MyDrive'):
    drive.mount('/content/drive')

# CHANGE THIS to your exact repo folder path
REPO_PATH = "/content/drive/MyDrive/Accenture Stream Training (SAP AI)"

# Go into repo
os.chdir(REPO_PATH)

# Add all changed files
os.system("git add .")

# Check if anything changed
changes = os.popen("git status --porcelain").read().strip()

if changes:
    msg = f"Auto sync {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
    os.system(f'git commit -m "{msg}"')
    os.system("git push")
    print("✅ Synced everything to GitHub")
else:
    print("✅ No changes to sync")
```
