## Task 1: Basic Python Operations

**Problem Statement:** Store one student's details (name, age, city, course, marks, attendance) in variables and print them in a clean, readable format.

**Approach:** Create simple Python variables of appropriate data types (string, int) for each attribute, then use an f-string to print them in the required layout.


```python
# Store individual student details in separate variables
student_name = "Rahul"          # string - student's name
age = 22                        # int - student's age
city = "Patna"                  # string - student's city
course = "Python"               # string - enrolled course
marks = 85                      # int - marks obtained out of 100
attendance_percentage = 92      # int - attendance percentage

# Print details in a clean, labeled format using f-strings
print(f"Student Name: {student_name}")
print(f"Age: {age}")
print(f"City: {city}")
print(f"Course: {course}")
print(f"Marks: {marks}")
print(f"Attendance: {attendance_percentage}%")
```

## Task 2: Conditional Statement

**Problem Statement:** Decide Pass/Fail based on marks (>=40 = Pass) and assign a Grade based on marks ranges (A+, A, B, C, Fail).

**Approach:** Use `if-elif-else` chains inside two functions — one to check pass/fail, another to map marks to a grade band. Testing top-down from the highest threshold ensures correct bucket assignment.


```python
def get_result(marks):
    return "Pass" if marks >= 40 else "Fail"

def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 75:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"

print(f"Result: {get_result(marks)}")
print(f"Grade: {get_grade(marks)}")
```

## Task 3: Loops

**Problem Statement:** Given a list of marks, print each value, count passes/fails, and compute highest, lowest, total, and average marks.

**Approach:** Iterate through the list once with a `for` loop, updating running counters (`passed_count`, `failed_count`, `total`) and tracking variables (`highest`, `lowest`) as we go, reinforcing manual loop logic instead of relying on built-ins.


```python
marks_list = [85, 92, 76, 88, 69, 91, 55, 78, 39, 95]

print("All marks:")
for m in marks_list:
    print(m)

passed_count = 0
failed_count = 0
for m in marks_list:
    if m >= 40:
        passed_count += 1
    else:
        failed_count += 1

highest = marks_list[0]
lowest = marks_list[0]
total = 0
for m in marks_list:
    if m > highest:
        highest = m
    if m < lowest:
        lowest = m
    total += m

average = total / len(marks_list)

print(f"\nPassed Students: {passed_count}")
print(f"Failed Students: {failed_count}")
print(f"Highest Marks: {highest}")
print(f"Lowest Marks: {lowest}")
print(f"Total Marks: {total}")
print(f"Average Marks: {average}")
```

## Task 5: Functions

**Problem Statement:** Wrap the result, grade, average, and attendance-status logic into reusable functions.

**Approach:** Define four standalone functions, each with a single responsibility, making the logic reusable later in Pandas (`.apply()`) and testable in isolation.


```python
def calculate_result(marks):
    """Return 'Pass' if marks >= 40, else 'Fail'."""
    return "Pass" if marks >= 40 else "Fail"

def calculate_grade(marks):
    """Map marks to a letter grade band."""
    if marks >= 90:
        return "A+"
    elif marks >= 75:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"

def calculate_average(marks_list):
    """Return the average of a list of marks."""
    return sum(marks_list) / len(marks_list)

def check_attendance_status(attendance):
    """Return 'Eligible' if attendance >= 75, else 'Low Attendance'."""
    return "Eligible" if attendance >= 75 else "Low Attendance"

print(calculate_result(85))
print(calculate_grade(85))
print(calculate_average([85, 92, 76, 88, 69]))
print(check_attendance_status(72))
```

## Task 6: Error Handling

**Problem Statement:** Accept marks as user input and handle non-numeric text, marks < 0, and marks > 100 without crashing.

**Approach:** Use `try/except` to catch `ValueError` on `float()` conversion, plus an `if` range check inside `try`. Loop with `while True` to re-prompt until valid. Since `input()` blocks notebook execution, we demonstrate with simulated inputs.


```python
def get_valid_marks():
    """Interactive version - repeatedly prompts until valid marks are entered."""
    while True:
        user_input = input("Enter marks (0-100): ")
        try:
            marks = float(user_input)
            if marks < 0 or marks > 100:
                print("Invalid input. Please enter marks between 0 and 100.")
                continue
            return marks
        except ValueError:
            print("Invalid input. Please enter marks between 0 and 100.")

def validate_marks(user_input):
    """Non-interactive version for demonstration/testing with fixed inputs."""
    try:
        marks = float(user_input)
        if marks < 0 or marks > 100:
            return "Invalid input. Please enter marks between 0 and 100."
        return f"Valid marks entered: {marks}"
    except ValueError:
        return "Invalid input. Please enter marks between 0 and 100."

test_inputs = ["85", "abc", "-5", "150", "76.5"]
for inp in test_inputs:
    print(f"Input: {inp!r} -> {validate_marks(inp)}")

# To take real interactive input, uncomment below:
# marks = get_valid_marks()
# print("Marks accepted:", marks)
```

## Dataset Setup

**Problem Statement:** Tasks 7 onward need a shared 10-record dataset.

**Approach:** Define the dataset once as a list of lists plus a header, reused across the remaining tasks to avoid duplication.


```python
import csv

students_data = [
    [1, "Rahul", 22, "Patna", "Python", 85, 92],
    [2, "Priya", 24, "Delhi", "Data Science", 92, 88],
    [3, "Aman", 21, "Kolkata", "Python", 76, 81],
    [4, "Neha", 23, "Mumbai", "Analytics", 88, 95],
    [5, "Rohan", 25, "Patna", "Python", 69, 72],
    [6, "Simran", 22, "Delhi", "Data Science", 91, 89],
    [7, "Vikash", 24, "Patna", "Analytics", 55, 68],
    [8, "Pooja", 21, "Kolkata", "Python", 78, 84],
    [9, "Arjun", 23, "Mumbai", "Data Science", 39, 60],
    [10, "Sneha", 22, "Delhi", "Python", 95, 97],
]

header = ["Student_ID", "Name", "Age", "City", "Course", "Marks", "Attendance_Percentage"]
```

## Task 8: CSV File Handling

**Problem Statement:** Create `students.csv`, then read it back and print each row.

**Approach:** Use `csv.writer` to write header + rows, `csv.reader` to read and iterate line by line.


```python
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(students_data)

print("students.csv created successfully.\n")

print("Reading students.csv:")
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

## Task 7: File Handling

**Problem Statement:** Write a text summary (totals, average, highest/lowest, pass/fail counts) into `student_summary.txt`, then read and display it.

**Approach:** Compute stats with list comprehensions and built-ins (`sum`, `max`, `min`), format into a multi-line string, and use standard file I/O with `with` statements.


```python
all_marks = [row[5] for row in students_data]

total_students = len(students_data)
avg_marks = sum(all_marks) / total_students
highest_marks = max(all_marks)
lowest_marks = min(all_marks)
passed_students = sum(1 for m in all_marks if m >= 40)
failed_students = total_students - passed_students

summary_text = f"""Student Performance Summary
Total Students: {total_students}
Average Marks: {round(avg_marks, 1)}
Highest Marks: {highest_marks}
Lowest Marks: {lowest_marks}
Passed Students: {passed_students}
Failed Students: {failed_students}
"""

with open("student_summary.txt", "w") as f:
    f.write(summary_text)

print("student_summary.txt created successfully.\n")

print("Reading student_summary.txt:")
with open("student_summary.txt", "r") as f:
    print(f.read())
```

## Task 9: NumPy Analysis

**Problem Statement:** Compute total, average, highest, lowest marks, and standard deviation using NumPy.

**Approach:** Convert marks list into `numpy.array` to use vectorized functions `np.sum`, `np.mean`, `np.max`, `np.min`, `np.std`.


```python
import numpy as np

marks_array = np.array(all_marks)

total_marks_np = np.sum(marks_array)
average_marks_np = np.mean(marks_array)
highest_marks_np = np.max(marks_array)
lowest_marks_np = np.min(marks_array)
std_marks_np = np.std(marks_array)

print(f"Total Marks: {total_marks_np}")
print(f"Average Marks: {round(average_marks_np, 2)}")
print(f"Highest Marks: {highest_marks_np}")
print(f"Lowest Marks: {lowest_marks_np}")
print(f"Standard Deviation: {round(std_marks_np, 2)}")
```

## Task 10: Pandas Data Analysis

**Problem Statement:** Load `students.csv` into a DataFrame and explore its structure.

**Approach:** Use `pd.read_csv()` then `.head()`, `.tail()`, `.shape`, `.columns`, `.info()`, `.describe()`.

## Task 11: Pandas Filtering

**Problem Statement:** Extract subsets: marks > 80, attendance < 75%, from Patna, Python course, failed students.

**Approach:** Use boolean indexing `df[condition]` to keep only matching rows.


```python
above_80 = df[df["Marks"] > 80]
low_attendance = df[df["Attendance_Percentage"] < 75]
from_patna = df[df["City"] == "Patna"]
python_students = df[df["Course"] == "Python"]
failed = df[df["Marks"] < 40]

print("Students who scored more than 80 marks:")
display(above_80)

print("\nStudents whose attendance is below 75%:")
display(low_attendance)

print("\nStudents from Patna:")
display(from_patna)

print("\nStudents enrolled in Python course:")
display(python_students)

print("\nStudents who failed:")
display(failed)
```

## Task 12: Pandas Sorting

**Problem Statement:** Sort by Marks (desc), Attendance (desc), Name (alphabetical).

**Approach:** Use `.sort_values(by=column, ascending=...)`.


```python
sorted_by_marks = df.sort_values(by="Marks", ascending=False)
sorted_by_attendance = df.sort_values(by="Attendance_Percentage", ascending=False)
sorted_by_name = df.sort_values(by="Name")

print("Sorted by Marks (descending):")
display(sorted_by_marks)

print("\nSorted by Attendance (descending):")
display(sorted_by_attendance)

print("\nSorted by Name (alphabetical):")
display(sorted_by_name)
```

## Task 13: Add New Columns

**Problem Statement:** Add `Result`, `Grade`, `Attendance_Status` columns.

**Approach:** Reuse rule functions, apply row-wise using `.apply()`.


```python
def result_col(marks):
    return "Pass" if marks >= 40 else "Fail"

def grade_col(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 75:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"

def attendance_status_col(att):
    return "Eligible" if att >= 75 else "Low Attendance"

df["Result"] = df["Marks"].apply(result_col)
df["Grade"] = df["Marks"].apply(grade_col)
df["Attendance_Status"] = df["Attendance_Percentage"].apply(attendance_status_col)

display(df)
```

## Task 14: Grouping and Aggregation

**Problem Statement:** City-wise/course-wise counts, avg marks by course, avg attendance by city, highest marks per course.

**Approach:** Use `.groupby()` with `.size()`, `.mean()`, `.max()`.


```python
city_counts = df.groupby("City").size()
course_counts = df.groupby("Course").size()
avg_marks_course = df.groupby("Course")["Marks"].mean()
avg_attendance_city = df.groupby("City")["Attendance_Percentage"].mean()
highest_marks_course = df.groupby("Course")["Marks"].max()

print("Students city-wise:")
print(city_counts)

print("\nStudents course-wise:")
print(course_counts)

print("\nAverage marks course-wise:")
print(avg_marks_course)

print("\nAverage attendance city-wise:")
print(avg_attendance_city)

print("\nHighest marks in each course:")
print(highest_marks_course)
```

## Task 15: Final Mini Project Report

**Problem Statement:** Produce a consolidated summary and save the final dataset as `final_student_performance_report.csv`.

**Approach:** Compute metrics from the enriched DataFrame, assemble a summary table, and export with `.to_csv()`.


```python
total_students_f = len(df)
average_marks_f = round(df["Marks"].mean(), 2)
highest_marks_f = df["Marks"].max()
lowest_marks_f = df["Marks"].min()
passed_students_f = (df["Result"] == "Pass").sum()
failed_students_f = (df["Result"] == "Fail").sum()
low_attendance_f = (df["Attendance_Status"] == "Low Attendance").sum()
above_80_f = (df["Marks"] > 80).sum()

final_report = pd.DataFrame({
    "Metric": [
        "Total Students", "Average Marks", "Highest Marks", "Lowest Marks",
        "Passed Students", "Failed Students", "Students with Low Attendance",
        "Students Scoring Above 80"
    ],
    "Value": [
        total_students_f, average_marks_f, highest_marks_f, lowest_marks_f,
        passed_students_f, failed_students_f, low_attendance_f, above_80_f
    ]
})

display(final_report)

df.to_csv("final_student_performance_report.csv", index=False)
print("\nfinal_student_performance_report.csv saved successfully.")
```
